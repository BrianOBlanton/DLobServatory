### NB to show getting JPL/ECCO2 data and making a basic 2-d spatial plot

```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
url='http://opendap.renci.org:1935/thredds/dodsC/oedata/SSH.1440x720.20140917.nc'
print url
```

```python
ds=nc.Dataset(url)
print 'Available variables are: ' +  str(ds.variables.keys())
```

```python
lon=ds.variables['LONGITUDE_T']
lat=ds.variables['LATITUDE_T']
X,Y = np.meshgrid(lon,lat)
```

```python
ssh=ds.variables['SSH']
sshd=ssh[0,:,:]
```

```python
fig = plt.figure(figsize=(24,12), dpi=144)
latmin = np.mean(lat)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_aspect(1.0/np.cos(latmin * np.pi / 180.0))
pc=plt.pcolormesh(X, Y, sshd, cmap=plt.cm.jet_r)
plt.title('Sea Surface Height, JPL/ECCO2, 17 Sep, 2014', fontsize=30)
plt.grid(True)
plt.xlim((-5,365))
plt.ylim((-92,92))
plt.tick_params(axis='both', which='major', labelsize=20)

# add colorbar
cbax = fig.add_axes([0.95, 0.1, 0.05, 0.8])
cb = plt.colorbar(pc, cax=cbax,  orientation='vertical')
cb.set_label('meters', fontsize=24)
cb.ax.tick_params(axis='both', which='major', labelsize=20)
```
