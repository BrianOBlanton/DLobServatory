```python
import netCDF4
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
%matplotlib inline
```

```python
nc=netCDF4.Dataset('./test.slosh.nc')
```

```python
lon=nc.variables['longitude']
lond=lon[:]-360
print np.min(lond), np.max(lond)
lat=nc.variables['latitude']
latd=lat[:]
print np.min(latd), np.max(latd)
latmn = np.mean(latd)
print latmn
```

```python
var=nc.variables['SURGE_0maboveground']
zeta=var[0,:,:]
print np.min(zeta), np.max(zeta)
```

```python
fig = plt.figure(figsize=(12,6), dpi=144)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_aspect(1.0/np.cos(latmn * np.pi / 180.0))

pc=pcolor(lond, latd, zeta, cmap=plt.cm.jet, vmin=0, vmax=10)
plt.title("SLOSH test", fontsize=20)
plt.grid(True)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.xlim((-80,-68))
plt.ylim(( 32, 45))

cbax = fig.add_axes()  #([0.95, 0.1, 0.05, 0.8])
cb = plt.colorbar(pc, cax=cbax,  orientation='vertical')
cb.set_label(SURGE_0maboveground.units, fontsize=14)
cb.ax.tick_params(axis='both', which='major', labelsize=12)
```
